```python
import os
import streamlit as st
from mistralai.client import MistralClient

# Set up Mistral AI Client
api_key = os.environ["Mdv0nFKHxWOtJYnTA0YVZGLHeBlzMLly"]
client = MistralClient(api_key=api_key)
model = "codestral-latest"

def refactor_code(code):
    # Define the prompt for refactoring
    prompt = f"Refactor the following Python code to improve readability and performance:\n\n{code}"
    response = client.completion(
        model=model,
        prompt=prompt,
        max_tokens=512,  # Adjust based on your needs
    )
    return response.choices[0].message.content

# Streamlit interface
st.title("AI-driven Code Refactoring Assistant")
code = st.text_area("Enter your code here:", height=300)
if st.button("Refactor Code"):
    if code:
        refactored_code = refactor_code(code)
        st.subheader("Original Code")
        st.code(code)
        st.subheader("Refactored Code")
        st.code(refactored_code)
    else:
        st.warning("Please enter some code to refactor.")
```
